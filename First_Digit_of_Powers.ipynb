This notebook shows graphically one very interesting mathematical fact that I encountered when talking to a friend of mine. 

Namely, the distribution of first digits of powers of numbers is not uniform as one would intuitively expect. In fact, it is uniform only for power p = 1, and for every other power, it is skewed towards the left, i.e. the most powers are starting with the digit 1, then 2 and so on to 9.

Even though initially I didn't expect this, it does indeed make sense. When we take a power of a number starting at a lower digit, the digit will usually stay low, while taking a power of a number starting at a higher digit, it will go to one with a lower digit as well. In other words, when we take a certain power of a number starting with 1 the result is much more likely to stay a number starting with 1, than if we did the same thing with a number starting with a 7.

Furthermore, the more we increase the power, the distribution becomes more skeweed to the left, i.e. to the smaller digits as visualized at the end of this notebook. 

```
import matplotlib.pylab as plt
```

```
# A function that returns the first digit from a number

def first_digit(n): 
    while n >= 10:
        n = n / 10;
    return int(n) 
```

```
# A function that initializes an empty dictionary

def initialize_empty_dic():
    Dict = {}
    for i in range (10):
        Dict[i] = 0;
    return Dict;
```

```
# Creating a figure on which we plot the results
fig = plt.figure(figsize =(20, 10)) 
# Choosing a power up to which we want to test the results
pow = 20;
for p in range(1, pow):
    d = initialize_empty_dic();
    for i in range (1, 10000):
        num = i**p;
        d[first_digit(num)] +=1;
    # add to the plot
    lists = sorted(d.items())
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    plt.plot(x, y, label = ('power = ' + str(p)))
    plt.legend(loc='upper right')
```
